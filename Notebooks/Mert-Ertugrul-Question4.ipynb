```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

## Geting The Data and Parsing The Number

```python
rdd=sc.textFile("datasets/SayisalLoto.txt")

rdd1= rdd.flatMap(lambda line: line.split("}{"))
rdd2=rdd1.map(lambda line :line.split(" "))
rdd3=rdd2.map(lambda line :line[41])
rdd4=rdd3.map(lambda line :line.replace(",",""))
rdd5=rdd4.map(lambda line :line.replace('"',""))

rdd5.take(5)
```

## Formating The Numbers

```python
def lit_to_tuple(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    a4 = int(arr[3])
    a5 = int(arr[4])
    a6 = int(arr[5])
    
    return [a1,a2,a3,a4,a5,a6]
```

```python
rdd6 = rdd5.map(lambda x: x.split('#'))
rdd6.take(5)
```

## Making a Cartesian

```python
rdd9 = rdd6
```

```python
rdd10 = rdd9.cartesian(rdd6)
rdd10.take(5)
```

# Function That Idendify, Given 2 Array Has 3 Common Numbers

## Wheather arrays has 3 common Numbers

```python
def hes(arr1,arr2):
    i = 0
    for a1 in arr1:
        for a2 in arr2:
            if a1 == a2:
                i += 1
    if i==3:
        return True
    else:
        return False
```

## List Common Numbers

```python
def hes_arr(arr1,arr2):
    i = 0
    arr = []
    for a1 in arr1:
        for a2 in arr2:
            if a1 == a2:
                i += 1
                arr.append(a1)
    if i==3:
        return arr
```

# Winners That Has 3 Common Numbers

```python
rdd10 = rdd10.filter(lambda x:hes(x[0],x[1])==True)
rdd10.take(10)
```

## With list of common numbers

```python
rdd10 = rdd10.map(lambda x: ( hes_arr(x[0],x[1]), x[0], x[1] ) )
rdd10.take(10)
```

## Sorting The Numbers

#### So that we can make a common numbers the key

```python
def sort_num(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    
    la = [a1, a2, a3]
    la.sort(reverse=True)
    las = []
    for a in la:
        las.append(str(a))
    return las
```

```python
def sort_num_str(arr):
    a1 = int(arr[0])
    a2 = int(arr[1])
    a3 = int(arr[2])
    
    la = [a1, a2, a3]
    la.sort(reverse=True)
    las = str(la[0])
    las = las + '-' + str(la[1])
    las = las + '-' + str(la[2])
    
    return las
```

Testing the sorting function

```python
sort_num_str(['05', '15', '47'])
```

# Winning Numbers RDD With 3 Common Numbers as The Key

```python
rdd10 = rdd10.map(lambda x: (sort_num_str(x[0]), x[1], x[2]))
rdd10.take(5)
```

## Just The 3 Common Numbers

```python
rdd11 = rdd10.map(lambda x: x[0])
rdd11.take(5)
```

# Counting The 3 Common Numbers

```python
rdd12 = rdd11.map(lambda x : (x,1))
rdd12.take(5)
```

```python
rdd12 = rdd12.reduceByKey(lambda x,y: x+y)
print(rdd12.collect())
```

### Sorting The Count

```python
rdd12 = rdd12.sortBy(lambda x: x[1], ascending=False)
rdd12.take(10)
```

## Since we made a cartesian RDD count the numbers twice, therefore we need to divide the count to 2

```python
rdd13 = rdd12.map(lambda x:(x[0], x[1]/2))
rdd13.take(50)
```
